# MBTI - Natural Language Processing

```python
import pandas as pd
import matplotlib.pyplot as plt
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize, RegexpTokenizer # 문장단위, 단어단위, 
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import Text
from nltk import FreqDist
from nltk.tag import pos_tag
```

```python
file = open('mbti.txt', encoding='utf8')
mbti = file.read()
```

```python
stopwords_mbti = set(line.strip() for line in open('stopwords_mbti.txt'))
```

```python
retokenize = RegexpTokenizer("[\w]+")
```

```python
text = Text(retokenize.tokenize(mbti))
text
```

```python
text.plot(25)
```

# Tokenizer

```python
mbti_tokens = pos_tag(retokenize.tokenize(mbti))
```

```python
stopwords_mbti = set(line.strip() for line in open('stopwords_mbti.txt'))
```

```python
nn_list = [t[0] for t in mbti_tokens if t[1] == "NN" and t[0] not in stopwords_mbti]
fd_mbti = FreqDist(nn_list)
```

```python
fd_mbti.N()
```

```python
fd_mbti.most_common(30)
```

# WordCloud

```python
from wordcloud import WordCloud
```

```python
wc = WordCloud(width=1920, height=1080, background_color='white', random_state=0)
plt.imshow(wc.generate_from_frequencies(fd_mbti))
plt.axis("off")
plt.show()
```
