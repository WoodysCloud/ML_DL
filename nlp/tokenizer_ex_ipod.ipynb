# tokenizer exercise: ipod reviews

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize, RegexpTokenizer
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import Text
```

```python
ipod = open('/Users/Woody/nltk_data/corpora/product_reviews_2/ipod.txt', encoding='utf8')
ipod = ipod.read()
# print(ipod)
```

```python
sent_tokenize(ipod[:1000])
```

```python
word_tokenize(ipod[:10])
```

```python
retokenize = RegexpTokenizer("[\w]+") #   .    X
retokenize.tokenize(ipod[:10])
```

```python
text = Text(retokenize.tokenize(ipod))
text
```

```python
text.plot(20)
```

```python
text.dispersion_plot(["iPod", "you", "best", "perfect", "happy", "good", "bad", "no"])
```

```python
from nltk import FreqDist
from nltk.tag import pos_tag

stopwords = ["Don", "of", "and", "t", "I", "m"]
ipod_tokens = pos_tag(retokenize.tokenize(ipod))
```

```python
ipod_tokens
```

```python
names_list = [t[0] for t in ipod_tokens if t[1] == "NN" and t[0] not in stopwords]
fd_names = FreqDist(names_list)
```

```python
fd_names.N()
```

```python
fd_names["iPod"]
```

```python
fd_names.freq("iPod")
```

```python
fd_names.most_common(5)
```
