data type이 중요. data type을 활용하여 data처리에 이용

```python
df3 = pd.DataFrame(data = [['hong', 10], ['park', 20]], columns = ['name', 'age'])
# 리스트 형태로 데이터프레임 생성
df3
```

```python
df3.dtypes
# object은 string을 말함
```

```python
df3['age'].astype(float) # dtype 변경
```

```python
df3
# CPU에서만 바꿈: 비파괴함수
# astype은 원본을 건들지 않으므로 비파괴함수
# RAM에서의 데이터는 그대로
```

```python
df3.dtypes
```

```python
df3['age'] = df3['age'].astype(float) # '='을 통해 RAM에 저장
# 위와 같이 변경(반영) 가능
df3
```

```python
df3['age'] = df3['age'].astype(int)
df3
```
