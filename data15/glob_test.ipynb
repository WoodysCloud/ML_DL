# 다수 파일 병합

```python
import pandas as pd
import glob
import re
from functools import reduce

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

from collections import Counter

import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
# 10개 파일 한꺼번에 불러오기, "asterisk" 사용
all_files = glob.glob("./data_list/my*.xls")
all_files
```

```python
# 저장할 리스트 생성
all_files_data = []
```

```python
# 10개 파일의 100개씩의 rows
for file in all_files:
    data_frame = pd.read_excel(file)
    # print(data_frame)
    all_files_data.append(data_frame)
print(len(all_files_data))
```

```python
print(all_files_data)
```

```python
# 행으로 합침
all_files_data_concat = pd.concat(all_files_data, axis=0, ignore_index=True)
all_files_data_concat
```

```python
# 제목 추출
all_title = all_files_data_concat['제목']
all_title
```

```python
# iterator data(반복자) - list, Series, tuple
```

```python
type(all_title)
```

```python
stopwords = set(stopwords.words("english")) # 영문에서 일반적으로 불필요한 단어들이 들어간 stopwords 객체 생성
lemma = WordNetLemmatizer() # 의미있는 단어(표제어) 추출 객체 생성
```

```python
words = []
```

```python
for title in all_title:
#     print(title)
    
    EnWords = re.sub(r"[^a-zA-Z]+", " ", str(title)) # 정규식
#     print(EnWords)
    
    EnWordsToken = word_tokenize(EnWords.lower())
#     print(EnWordsToken)
    
    EnWordsTokenStop = [w for w in EnWordsToken if w not in stopwords] # 익명함수, 클로저. 아래함수와 같은 결과 출력 코드    
    '''
    def call (EnWordsToken):
        result = []
        for w in EnWordsToken:
            if w not in stopwords:
                result.apped(w)
        return result
    '''
#     print(EnWordsTokenStop)
    
    EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop]
    '''
    def call2(EnWordsTokenStop):
        result = []
        for w2 in EnWordsTokenStop:
            result.append(lemma.lemmatize(w2))
        return result
    '''
#     print(EnWordsTokenStopLemma)

    words.append(EnWordsTokenStopLemma)
```

```python
words # 2차원 => reduce 함수 사용으로 1차원으로 변환
words[:10]
```

```python
words2 = reduce(lambda x, y: x + y, words)
print(type(words2[:10]))
words2[:10]
```

```python
count = Counter(words2)
count
```

```python
word_count = dict()
```

```python
count.most_common(5)
```

```python
for tag, counts in count.most_common(50):
    if len(str(tag)) > 1: # 1글자인 단어 필터링
        word_count[tag] = counts
print(word_count)
```

```python
# 'big', 'data' 제거
del word_count['big']
del word_count['data']
```
